```python
pyautogui.PAUSE = 1

#Comando pyautogui
#pyautogui.hoykey -> conjunto de teclas
#pyautogui.whrite -> escrever um texto
#pyautogui.press -> apertar 1 tecla


# Passo 1: Entrar no sistema (no nosso caso, entrar no link)
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl","v")
pyautogui.press("enter")

#demora alguns segundos
time.sleep(4)

#Passo 2: Navegar no sistema e encontrar a base dados(entrar na pasta exportar)
# Para achar a coordenada do local do click é preciso usar o pyautogui.position()
pyautogui.click(x=432, y=309, clicks=2)
time.sleep(2)

#Passo 3: Exportar/Fazer Download da Base de Dados
pyautogui.click(x=398, y=368) #clicar no arquivo
pyautogui.click(x=1160, y=193) #clicar nos 3 ponto
pyautogui.click(x=949, y=554, button="right") #clicar para fazer download
time.sleep(5)

#Passo 4: Importar a base de dados para o Python
tabela = pd.read_excel(r"C:\Users\luckh\Downloads\Vendas - Dez.xlsx")
display(tabela)

#Passo 5: Calcular os indicadores
#faturamento - soma da coluna 'valor final'
faturamento = tabela["Valor Final"].sum()
#quantidade -  soma da coluna 'quantidade'
quantidade = tabela["Quantidade"].sum()
print(f"Faturamento: {faturamento}")
print(f"Quantidade: {quantidade}")

#Passo 6: Enviar um email para a diretoria com o relatório
#abrir o email (link: https://mail.google.com/mail/u/0/#inbox)
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl","v")
pyautogui.press("enter")
time.sleep(2)
#clicar no escrever
pyautogui.click(x=131, y=192)
time.sleep(2)
#digita o destinatário
pyautogui.write("lucas.gabryell@souunit.com.br") #testar com email que voce tenha acesso para saber se chegou
pyautogui.press('tab')
pyautogui.press('tab')
#escreve o assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl","v")
pyautogui.press('tab')
#escreve o corpo do email
texto = f"""
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Abs
Lucas Romão
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl","v")
#envia o email
pyautogui.hotkey("ctrl", "enter")
```
